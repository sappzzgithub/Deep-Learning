```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
# Define augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=30,     # Rotate images by 30 degrees
    width_shift_range=0.2, # Shift image width by 20%
    height_shift_range=0.2,# Shift image height by 20%
    shear_range=0.2,       # Shear angle
    zoom_range=0.2,        # Zoom in/out by 20%
    horizontal_flip=True,  # Flip images horizontally
    vertical_flip=True,    # Flip images vertically
    fill_mode='nearest'    # Fill in missing pixels with nearest value
)
```

```python
#load the original img
img_path=r'/Users/sakshizanjad/Desktop/SEM 2/DL/download.jpeg'
img= load_img(img_path)  # Convert to array
x = img_to_array(img) # Expand dimensions
```

```python
#reshape img to 4d tensor
img=np.expand_dims(x,axis=0)
```

```python
#generate agumented img
agumented_images=[]
for batch in datagen.flow(img,batch_size=1):
  agumented_images.append(batch[0].astype(np.uint8))#for display
  if len(agumented_images)>=5:
    break
```

```python
#display the original agumented images
plt.figure(figsize=(10,5))
for i,img in enumerate(agumented_images):
  plt.subplot(1,5,i+1)
  plt.imshow(img)
  plt.title(f"Agumented images{i}")
  plt.axis('off')
  plt.tight_layout()
plt.show()
```
